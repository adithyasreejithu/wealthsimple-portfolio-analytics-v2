```python
import pandas as pd 
from pathlib import Path
file = Path("C:\\Projects\\Python\\wealthsimple-portfolio-analytics-v2\\next.csv")

required_cols = {
    "BUY": ["date", "transaction", "ticker_id", "quantity", "execDate", "debit"],
    "SELL": ["date", "transaction", "ticker_id", "quantity", "execDate"],
    "LOAN": ["date", "transaction", "ticker_id", "quantity", "execDate"],
    "RECALL": ["date", "transaction", "ticker_id", "quantity", "execDate"],
    "DIV": ["date", "transaction", "ticker_id", "execDate", "credit"],
    "NRT": ["date", "transaction", "execDate", "debit"],
    "FPLINT": ["date", "transaction", "credit"],
}

df = pd.read_csv(file)
df.reset_index(inplace=True)
```

```python
invalid_indexes = []

for trans, cols in required_cols.items():
    subset = df.loc[df["transaction"].eq(trans)]

    missing_required = subset[cols].isna().any(axis=1)

    invalid_indexes.extend(subset.index[missing_required])

invalid_rows = df.loc[invalid_indexes]
cleaned = df.drop(index=invalid_indexes).reset_index(drop=True)
```
