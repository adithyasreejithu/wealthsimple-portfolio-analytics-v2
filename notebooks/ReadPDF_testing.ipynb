```python
data = tables[3].df
data.head(50)
# data.to_csv("test.csv")
```

```python
data = data.drop([0,1])
data.rename(columns={0 : "date", 1: "transaction", 2: "info", 3: "debit", 4: "credit", 5:"balance"},inplace=True)
data = data.reset_index(drop=True)
```

```python
data.head(100)
```

```python
cond = data['date']== ""
data[cond]
```

```python
data['date'] = data['date'].replace('',None)
data['crit'] = data['date'].notna().cumsum()
```

```python
data.groupby("crit", as_index=False).agg({
"date": "first",
"transaction": "first",
"info": lambda x: " ".join(x.dropna().astype(str).str.strip()),
"debit": "first",
"credit": "first",
"balance": "first"
}).drop(columns="crit")
```
